```python
###CLEANING OF DATA###

'''Since our data stored in variable 'a' (olympics.csv) is quite clean
        so, we will use new dataset that is usually more dirty'''

e=pd.read_csv('dirtydata.csv')
e  #printing the dirty data
    
```

```python
# using isnull() method to check the null values (i.e zeroes)

'''it will return true if there exists a null value and 
false if there exists no null value in a complete column'''

e.isnull()
```

```python
e.isnull().sum() #It will us about the total number of null values in respective columns.
```

```python
'''now as we know whether there is a null value or not,
   we can wasily remove such rows with the help of .dropna() method'''

'''either we can use a new variable or use the inplace = True argument:
                to override the original data'''

f=e.dropna() 
f
```

```python
f.isnull().sum()  # now we can see that all the null values are removed
```

```python
# NOTE-we can use df.fillna('any value', inplace = True) to replace all the null values with this

# all the null values will get replaced by the entered value.

# df["Calories"].fillna(130, inplace = True) this way we can only replace null values within a specific column.
```

```python
# Cleaning wrong format
# as date was wrong in row 26 so we converted the whole column to dates.
# This was possible with the help of datetime method
# f['Date']=pd.to_datetime(f['Date'])
f
```

```python
# Cleaning of data with wrong values
# There are 3 ways to do so

''' (1) One way to fix wrong values is to replace them with something else.
       here i am giving higher and wrong value, so i may show use of other two methods '''

f.loc[7,'Duration']=110


''' (2) For small data sets you might be able to replace the wrong data one by one, but not for big data sets.
     In such cases loop through all values in the "Duration" column. If the value is higher than 120, set it to 120 '''

for x in f.index:
  if f.loc[x, "Duration"] > 110:
    f.loc[x, "Duration"] =70
    
    
''' (3) getting rid of wrong values by completeley removing them '''
for i in f.index:
    if f.loc[i,'Duration']>60:
        f.drop(i,inplace=True) 
```

```python
# we can see below that row 7 is completely removed.
f
```

```python
# Removing duplicates
# as we can encounter duplicate values in a dataset (look for row 11 and 12 in the above dataset)
# To discover duplicates, we can use the duplicated() method.
# The duplicated() method returns a Boolean values for each row

f.duplicated()

#By taking a look at our test data set, we can assume that row 11 and 12 are duplicates.
```

```python
# To remove duplicates, use the drop_duplicates() method.
g=f.drop_duplicates()
g

#in the output data we can see that 12yh row is removed because of being duplicate.
```

```python
#corr() method 

'''The Result of the corr() method is a table with a lot of numbers
that represents how well the relationship is between two columns.

The number varies from -1 to 1.

1 means that there is a 1 to 1 relationship (a perfect correlation),
and for this data set, each time a value went up in the first column, the other one went up as well.

0.9 is also a good relationship, and if you increase one value, 
the other will probably increase as well.

-0.9 would be just as good relationship as 0.9,
but if you increase one value, the other will probably go down.

0.2 means NOT a good relationship,
meaning that if one value goes up does not mean that the other will.'''
```
